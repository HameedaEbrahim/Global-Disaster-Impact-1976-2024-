## Fix the error in the Date " This done by search"

```python
import calendar

def fix_invalid_dates(df, year_col, month_col, day_col):
    for idx, row in df.iterrows():
        y = int(row[year_col])
        m = int(row[month_col])
        d = int(row[day_col])

        # max days in this month/year
        max_day = calendar.monthrange(y, m)[1]

        # if day is higher than allowed, correct it
        if d > max_day:
            df.at[idx, day_col] = max_day

    return df

# Apply it
disaster_df1 = fix_invalid_dates(
    disaster_df1,
    'Start Year', 'Start Month', 'Start Day'
)
```
